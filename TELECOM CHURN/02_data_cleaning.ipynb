```python

import pandas as pd
import numpy as np
import os

# 1. Chargement du dataset original
raw_data_path = "c:/Users/osee0/OneDrive/Bureau/TELECOM CHURN/data/WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(raw_data_path)
print(f"🔹 Dimension initiale du dataset : {df.shape[0]} lignes, {df.shape[1]} colonnes/n")
```

```python
df.head()
```

```python
# 2. Correction du piège de la colonne 'TotalCharges'
# On remplace les espaces vides cachés par des vrais NaN, puis on force le type en numérique
df['TotalCharges'] = df['TotalCharges'].replace(r'^/s*$', np.nan, regex=True)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# 3. Traitement logique des valeurs manquantes (Imputation Métier)
# Inspectons les lignes où 'TotalCharges' est désormais NaN
missing_total_charges = df[df['TotalCharges'].isnull()]
print(f"⚠️ Nombre de vraies valeurs manquantes trouvées dans 'TotalCharges' : {len(missing_total_charges)}")

```

```python
# Logique métier : Si le client a une fidélité (tenure) de 0 mois, c'est qu'il vient de souscrire.
# Ses charges totales sont donc légitimement de 0.
print("💡 Analyse métier : Vérification de la fidélité (tenure) pour ces profils...")
print(missing_total_charges['tenure'].value_counts())
```

```python

# Imputation : On remplace ces NaN par 0
df['TotalCharges'] = df['TotalCharges'].fillna(0.0)
print("✅ Imputation réussie : Valeurs manquantes fixées à 0.0 pour les nouveaux clients./n")

```

```python
# 4. Nettoyage de la cohérence des chaînes de caractères
# Certaines colonnes ont des modalités redondantes (ex: 'No internet service' au lieu de 'No')
# Pour l'instant on garde la structure brute, mais on vérifie les doublons d'identifiants
duplicate_counts = df['customerID'].duplicated().sum()
print(f"🔹 Nombre de doublons d'identifiants (customerID) : {duplicate_counts}")
```

```python

# 5. Sauvegarde du Dataset Propre (Processed)
# Création du dossier s'il n'existe pas
output_dir = "c:/Users/osee0/OneDrive/Bureau/TELECOM CHURN/data"
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, "processed_telco.csv")
df.to_csv(output_path, index=False)
print(f"/n🚀 Dataset nettoyé exporté avec succès vers : {output_path}")
```
