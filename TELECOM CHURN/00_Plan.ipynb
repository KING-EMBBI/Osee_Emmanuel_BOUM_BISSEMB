### PHASE 1 — Business Understanding & Cadrage Sectoriel

**Objectif :** Poser le problème en tant que consultant, et non en tant que simple codeur.

* **Étude des KPI Télécom clés :**
* **ARPU (Average Revenue Per User) :** Revenu moyen par utilisateur.
* **Churn Rate :** Taux d'attrition mensuel/annuel.
* **CLV (Customer Lifetime Value) :** Valeur financière d'un client sur toute sa durée de vie.
* **CAC (Customer Acquisition Cost) :** Ce qu'il en coûte pour acquérir un nouveau client vs le retenir.


* **Note de contextualisation (Le "Plus" Afrique) :** Rédiger un paragraphe expliquant que bien que le dataset soit basé sur un modèle *Postpaid* (abonnements/factures), la démarche logique s'applique parfaitement aux flottes d'entreprises (B2B) ou aux segments d'abonnements des opérateurs locaux (Orange, MTN, etc.).
* **Livrable :** `01_business_understanding.ipynb` posant les bases économiques du projet.

### PHASE 2 — Data Cleaning & Intégrité

**Objectif :** Garantir une base saine pour la modélisation statistique.

* **Correction du typage :** Forcer la colonne `TotalCharges` en numérique (gestion des espaces vides qui bloquent le type).
* **Analyse des valeurs manquantes :** Identifier si les données manquantes dans `TotalCharges` correspondent à des nouveaux clients (`tenure == 0`). Si oui, imputer à 0 (logique métier).
* **Vérification de la qualité :** Recherche active de doublons (`customerID`).
* **Livrable :** `02_data_cleaning.ipynb` + sauvegarde du fichier `data/processed/processed_telco.csv`.

### PHASE 3 — Exploratory Data Analysis (EDA) & Tests Statistiques

**Objectif :** Extraire des *insights* visuels et valider des hypothèses par les statistiques.

* **Analyse Univariée :** Distribution du Churn (calculer le déséquilibre exact des classes), de la fidélité (`tenure`) et des frais mensuels (`MonthlyCharges`).
* **Analyse Bivariée & Hypothèses Métier :**
* *Hypothèse Contract :* Est-ce que le contrat "Month-to-month" sur-génère du churn ?
* *Hypothèse Services :* Est-ce que l'absence de support technique (`TechSupport`) ou de sécurité en ligne augmente le risque de départ ?


* **Validation Statistique :**
* Test du Chi-deux ($\chi^2$) entre le Churn et les variables catégorielles (Contrat, Services).
* Test ANOVA ou T-Test pour comparer les `MonthlyCharges` entre les clients restés et partis.


* **Graphiques Premium :** Boxplots/Violin plots pour les charges, Stacked Bar Plots pour les types de contrats.
* **Livrable :** `03_exploratory_analysis.ipynb` contenant des conclusions écrites en clair (Ex: *"Les clients sans TechSupport ont un taux de churn 2.4 fois supérieur"*).

### PHASE 4 — Feature Engineering & Robustesse Statistique (Anti-Leakage)

**Objectif :** Préparer les données pour les algorithmes en évitant les biais de méthodologie.

* **Création de variables (Features) :**
* `tenure_groups` : Segmentation (0-12 mois, 12-24 mois, etc.).
* `total_services` : Somme des services activés par le client (Internet, Sécurité, Streaming...).


* **Séparation des données (Train/Test Split) :** **Crucial !** Faire le split (70/30 ou 80/20) *avant* toute étape d'encodage ou de scaling pour éviter les fuites d'information (*Data Leakage*).
* **Preprocessing Pipeline :**
* Encodage des variables catégorielles (One-Hot Encoding).
* Standardisation (StandardScaler) des variables numériques (`MonthlyCharges`, `TotalCharges`).


* **Gestion du déséquilibre des classes (Imbalance) :** Prévoir deux approches dans ton code pour comparer : l'utilisation des poids de classes (`class_weight='balanced'`) vs l'algorithme SMOTE (uniquement sur le train set).
* **Livrable :** `04_feature_engineering.ipynb`.

### PHASE 5 — Modélisation & Sélection Arbitrée par le Métier

**Objectif :** Entraîner les modèles et choisir le meilleur en fonction d'un compromis Business.

* **Modèles à confronter :**
1. **Baseline :** Régression Logistique (pour l'interprétation pure des coefficients et des *Odds Ratios*).
2. **Ensemble Learning :** Random Forest (robuste, moins sensible à l'overfitting).
3. **Boosting :** XGBoost ou LightGBM (recherche de la performance brute).


* **Validation :** Cross-Validation Stratifiée (*Stratified K-Fold*) pour maintenir la proportion de churn dans chaque sous-ensemble.
* **Arbitrage des Métriques :**
* Suivi de l'AUC-ROC et de la courbe Precision-Recall (PR-AUC).
* **Le choix du compromis :** Ne pas viser uniquement le *Recall* (qui détruirait la rentabilité en ciblant trop de faux positifs), mais utiliser le **F1-Score** ou un **F-beta score (ex: $\beta=2$)** pour donner une priorité calculée au Recall tout en contrôlant la Précision.


* **Livrables :** `05_modeling.ipynb` + scripts Python `src/preprocessing.py`, `src/train.py`, `src/evaluate.py`.

### PHASE 6 — Explainability (XAI) & Transparence Algorithmique

**Objectif :** Ouvrir la "boîte noire" du meilleur modèle pour donner des leviers d'action au marketing.

* **Analyse Globale avec SHAP :** Quelles sont les variables qui poussent le plus à l'attrition sur l'ensemble de la base clients ?
* **Analyse Locale (Micro) :** Prendre 2 profils types (un client fidèle à faible risque et un client récent à très haut risque) et afficher le graphique de force (*Waterfall/Force Plot*) SHAP pour expliquer précisément pourquoi le modèle s'alarme pour ce client spécifique.
* **Livrable :** `06_model_interpretation.ipynb` avec des visualisations prêtes à être intégrées dans le rapport final.

### PHASE 7 — Simulation Financière & Calcul du ROI

**Objectif :** Transformer les probabilités de churn en valeur monétaire. C'est l'étape reine du plan.

* **Définition des paramètres de simulation :**
* *Coût du churn :* Perte de l'ARPU annuel du client.
* *Coût de l'action marketing :* Prix d'un package de rétention (ex: 15% de réduction ou un bonus data pendant 6 mois).
* *Taux d'acceptation :* Hypothèse réaliste du pourcentage de clients à risque qui acceptent l'offre et restent (ex: 20%).


* **Calcul du gain net (ROI) :** Comparer le coût de la situation sans modèle (perte totale des churners) vs la situation avec ton modèle (coût des campagnes de rétention + coût des faux positifs vs pertes évitées).
* **Analyse de scénarios :** Proposer trois scénarios : Pessimiste, Réaliste, Optimiste.
* **Livrable :** `07_business_simulation.ipynb`.

### PHASE 8 — Le Dashboard Exécutif Interactif

**Objectif :** Donner un outil d'aide à la décision visuel pour la direction.

* **Technologie recommandée :** **Power BI** (pour une immersion "Corporate" très recherchée par Orange/MTN) ou **Streamlit/Shiny** (pour valoriser tes compétences d'ingénierie data).
* **Vue d'ensemble :** KPI du taux de churn global, perte financière estimée, curseur de sélection du niveau de risque, et liste des clients prioritaires à contacter par le service client.
* **Livrable :** Fichier de dashboard dans le dossier `dashboard/`.

* **README.md GitHub :** Doit inclure un résumé percutant, l'architecture du projet, les performances du modèle et surtout, un **GIF animé** ou des captures d'écran de ton dashboard.
