 <strong> 1. Contexte Économique & Enjeux Sectoriels</strong> 
</center>
<div style="background-color: #ffffff; padding: 30px 40px; color: #2c3e50; font-family: 'Calibri', 'Segoe UI', Arial, sans-serif; font-size: 16px; line-height: 1.6; text-align: justify; box-shadow: 0 4px 15px rgba(0,0,0,0.05); border-radius: 4px; margin: 20px 0; border-left: 5px solid #2e7d32;">
    <p style="margin-bottom: 20px; text-indent: 30px;">
        Dans l'industrie des télécommunications, caractérisée par une concurrence féroce et une saturation progressive des marchés, la perte de clients ou <strong>churn</strong> (attrition) représente l'un des défis majeurs pour la rentabilité des opérateurs. D'un point de vue purement financier, les études sectorielles démontrent de manière constante qu'acquérir un nouveau client coûte entre <strong>5 et 7 fois plus cher</strong> que de retenir un client existant. Le Coût d'Acquisition Client (CAC) englobe des dépenses massives en marketing, en publicité, en offres de bienvenue et en commissions de distribution. À l'inverse, la fidélisation permet de maximiser la Valeur Vie Client (<em>Customer Lifetime Value - CLV</em>) et de stabiliser l'ARPU (<em>Revenu Moyen par Utilisateur</em>). Une réduction, même minime, du taux de churn de l'ordre de 1% à 2% peut se traduire par des millions de FCFA d'économies annuelles pour un opérateur de premier plan.
    </p>
    <p style="background-color: #f8f9fa; padding: 15px; border-radius: 4px; font-size: 15px; color: #555555; font-style: italic; margin-top: 25px; border-left: 3px solid #6c757d;">
        <strong>Note de contextualisation locale :</strong> Bien que le marché télécom d'Afrique subsaharienne soit historiquement dominé par le modèle <em>Prepaid</em> (cartes de recharge et forfaits ponctuels), le segment <em>Postpaid</em> (abonnements mensuels, flottes B2B d'entreprises et offres Internet résidentielles Fibre/Box) constitue le moteur à haute valeur ajoutée des opérateurs. La méthodologie développée ici cible la sécurisation de ces revenus contractuels stratégiques.
    </p>
</div>

<center style="background-color: white; padding:18px; color: #1a252f; font-family:'Tw Cen MT', sans-serif; font-size: 20px;">
 <strong> 2. Objectif du Projet : Vers une Rétention Proactive</strong> 
</center>
<div style="background-color: #ffffff; padding: 30px 40px; color: #2c3e50; font-family: 'Calibri', 'Segoe UI', Arial, sans-serif; font-size: 16px; line-height: 1.6; text-align: justify; box-shadow: 0 4px 15px rgba(0,0,0,0.05); border-radius: 4px; margin: 20px 0; border-left: 5px solid #2e7d32;">
    <p style="margin-bottom: 20px; text-indent: 30px;">
        L'objectif fondamental de ce projet est d'opérer un changement de paradigme au sein de la direction marketing : passer d'une stratégie de rétention réactive à une stratégie proactive.
    </p>
    <ul>
        <li><strong>Approche Réactive (Traditionnelle) :</strong> L'opérateur intervient uniquement lorsque le client manifeste explicitement son intention de partir (appel au service client, demande de résiliation). À ce stade, le client est souvent déjà psychologiquement acquis à la concurrence, et les offres de rétention de dernière minute s'avèrent coûteuses et peu efficaces.</li>
        <li><strong>Approche Proactive (Data-Driven) :</strong> Grâce aux algorithmes de Machine Learning, nous analysons les signaux faibles et les comportements d'utilisation pour identifier les clients à haut risque de churn avant qu'ils ne prennent leur décision. Cela permet aux équipes marketing de déclencher des campagnes de rétention ciblées, personnalisées et optimales sur le plan budgétaire.</li>
    </ul>
</div>

<center style="background-color: white; padding:18px; color: #1a252f; font-family:'Tw Cen MT', sans-serif; font-size: 20px;">
 <strong> 3. Définition de la Problématique Analytique</strong> 
</center>
<div style="background-color: #ffffff; padding: 30px 40px; color: #2c3e50; font-family: 'Calibri', 'Segoe UI', Arial, sans-serif; font-size: 16px; line-height: 1.6; text-align: justify; box-shadow: 0 4px 15px rgba(0,0,0,0.05); border-radius: 4px; margin: 20px 0; border-left: 5px solid #2e7d32;">
    <p style="margin-bottom: 20px; text-indent: 30px;">
        Pour résoudre ce problème business, nous devons le traduire en une problématique de Data Science. L'enjeu consiste à bâtir un modèle mathématique capable d'estimer la probabilité qu'un client donné quitte l'opérateur au cours du mois prochain. Sur le plan algorithmique, il s'agit d'un problème de classification binaire, où la variable cible à prédire (le Target) possède deux états exclusifs :
        <br><br>
        • <strong>Y = 1</strong> : Le client va churner (résiliation de son profil).<br>
        • <strong>Y = 0</strong> : Le client reste fidèle (rétention active).<br><br>
        Cette prédiction s'appuiera sur un vecteur de caractéristiques (features) regroupant l'historique démographique du client, les modalités de son contrat (durée, mode de paiement), sa structure de coûts (MonthlyCharges, TotalCharges) et son niveau d'engagement envers les services additionnels de l'opérateur.
    </p>
</div>

<center style="background-color: white; padding:18px; color: #1a252f; font-family:'Tw Cen MT', sans-serif; font-size: 20px;">
 <strong> 4. Hypothèses Métier à vérifier (Feuille de route pour l'EDA)</strong> 
</center>
<div style="background-color: #ffffff; padding: 30px 40px; color: #2c3e50; font-family: 'Calibri', 'Segoe UI', Arial, sans-serif; font-size: 16px; line-height: 1.6; text-align: justify; box-shadow: 0 4px 15px rgba(0,0,0,0.05); border-radius: 4px; margin: 20px 0; border-left: 5px solid #2e7d32;">
    <p style="margin-bottom: 20px; text-indent: 30px;">
        Avant de concevoir le modèle prédictif, la phase d'Analyse Exploratoire des Données (EDA) devra valider ou infirmer statistiquement les hypothèses stratégiques suivantes :
        <br><br>
        <strong>Hypothèse 1 (Facteur d'engagement) :</strong><br>
        Les clients engagés sur des contrats de court terme (Month-to-month) présentent une volatilité et un taux de churn mécaniquement plus élevés que ceux sécurisés par des engagements contractuels de 1 ou 2 ans.
        <br><br>
        <strong>Hypothèse 2 (Facteur Qualité/Prix) :</strong><br>
        Les clients subissant des charges mensuelles élevées (MonthlyCharges) tout en étant privés de services de support technique (TechSupport) disposent d'un ratio "coût/valeur perçue" défavorable, augmentant drastiquement leur probabilité de départ.
    </p>
</div>

<center style="background-color: white; padding:18px; color: #1a252f; font-family:'Tw Cen MT', sans-serif; font-size: 20px;">
 <strong> 5. Quelques rappels pratiques</strong> 
</center>
<div style="background-color: #ffffff; padding: 30px 40px; color: #2c3e50; font-family: 'Calibri', 'Segoe UI', Arial, sans-serif; font-size: 16px; line-height: 1.6; text-align: justify; box-shadow: 0 4px 15px rgba(0,0,0,0.05); border-radius: 4px; margin: 20px 0; border-left: 5px solid #2e7d32;">
    <p style="background-color: #f8f9fa; padding: 15px; border-radius: 4px; font-size: 15px; color: #555555; font-style: italic; border-left: 3px solid #6c757d;">
        <strong>CONCEPT CLÉ — La "boîte noire" (Black Box) :</strong> Fait référence à un modèle de Machine Learning complexe (comme XGBoost ou les réseaux de neurones) dont les mécanismes internes de décision sont mathématiquement trop opaques pour être interprétés directement par un humain.
        <br><br>
        <strong>Le principe de la boîte noire :</strong> L'algorithme génère des prédictions d'une précision chirurgicale en entrée (X), mais le chemin logique qui mène au résultat en sortie (y_pred) reste complexe à isoler pour l'utilisateur.
        <br><br>
        C'est pour briser ce principe et redonner le contrôle aux équipes métiers que nous intégrons le framework <strong>SHAP (Shapley Additive exPlanations)</strong>, une méthode issue de la théorie des jeux qui permet de mesurer l'impact exact et transparent de chaque variable sur la décision finale du modèle.
    </p>
